```python
# Convert date column to datetime
df['Order.Date'] = pd.to_datetime(df['Order.Date'], errors='coerce')
```

```python
# Convert category/region to categorical
df['Category'] = df['Category'].astype('category')
df['Region'] = df['Region'].astype('category')
```

```python
# Convert numerical columns
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')
```

```python
# Profit Margin
df['Profit Margin %'] = (df['Profit'] / df['Sales']) * 100
```

```python
# Discount Impact
df['Revenue After Discount'] = df['Sales'] * (1 - df['Discount'])
```

```python
# Month and Year for Time Analysis
df['Year'] = df['Order.Date'].dt.year
df['Month'] = df['Order.Date'].dt.month_name()
```

```python
# Example: Remove extreme sales values
q1 = df['Sales'].quantile(0.25)
q3 = df['Sales'].quantile(0.75)
iqr = q3 - q1
lower_limit = q1 - 1.5 * iqr
upper_limit = q3 + 1.5 * iqr
df = df[(df['Sales'] >= lower_limit) & (df['Sales'] <= upper_limit)]
```
