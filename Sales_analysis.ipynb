```python
print(data.dtypes)
for c in ['Sales','Profit','Quantity','Discount']:
    print(c, "unique types:", data[c].apply(type).unique()[:5])
# convert if necessary
data['Sales']  = pd.to_numeric(data['Sales'], errors='coerce')
data['Profit'] = pd.to_numeric(data['Profit'], errors='coerce')
data['Quantity'] = pd.to_numeric(data['Quantity'], errors='coerce', downcast='integer')
data['Discount'] = pd.to_numeric(data['Discount'], errors='coerce')
```

```python
print("Total Sales:", data['Sales'].sum(), "Total Profit:", data['Profit'].sum())
```

Standardize & correct (clean text, dates, codes)


```python

data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')
data['Ship Date']  = pd.to_datetime(data['Ship Date'], errors='coerce')
data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month
data['MonthName'] = data['Order Date'].dt.strftime('%b')
data['YearMonth'] = data['Order Date'].dt.to_period('M').astype(str)  # "2020-03"
data['Quarter'] = data['Order Date'].dt.to_period('Q').astype(str)
```

Trim & normalize strings

Remove leading/trailing spaces; unify case (title case for cities).

```python
text_cols = ['Customer Name','City','State','Region','Category','Sub-Category','Product Name']
for c in text_cols:
    data[c] = data[c].astype(str).str.strip()
data['City'] = data['City'].str.title()
data['State'] = data['State'].str.title()
data['Category'] = data['Category'].str.title()
```

post code


```python
data['Postal Code'] = data['Postal Code'].astype(str).str.strip()
```

```python
bad_discounts = data[(data['Discount'] < 0) | (data['Discount'] > 1)]
print("bad discounts:", bad_discounts.shape[0])
```

```python
neg_sales = data[data['Sales'] < 0]
neg_profit = data[data['Profit'] < 0]
```

```python
print(data['Category'].unique())
print(data['Sub-Category'].unique())
```

Enrichment / Feature engineering

Days to ship

```python
data['DaysToShip'] = (data['Ship Date'] - data['Order Date']).dt.days
```

Profit Margin

```python
data['ProfitMargin'] = data['Profit'] / data['Sales']
data.loc[data['Sales']==0, 'ProfitMargin'] = 0
```

Order totals & Average Order Value (AOV)

```python
data['OrderTotal'] = data.groupby('Order ID')['Sales'].transform('sum')
# If you need per-customer AOV:
cust_aov = data.groupby('Customer ID')['OrderTotal'].mean().reset_index().rename(columns={'OrderTotal':'AOV'})
data = data.merge(cust_aov, on='Customer ID', how='left')
```

Top N flags / ranks (for top products/customers)

```python
product_sales = data.groupby('Product Name')['Sales'].sum().reset_index().sort_values('Sales',ascending=False)
product_sales['Rank'] = product_sales['Sales'].rank(method='dense', ascending=False)
# merge rank if needed
data = data.merge(product_sales[['Product Name','Rank']], on='Product Name', how='left')
```

```python
data['ProfitPerUnit'] = data['Profit'] / data['Quantity'].replace(0, pd.NA)
# Avg deal size per order:
data['AvgDealSize'] = data.groupby('Order ID')['Sales'].transform('mean')
```

```python
monthly = data.groupby(['Region','YearMonth'])['Sales'].sum().reset_index()
avg_monthly = monthly.groupby('Region')['Sales'].mean().reset_index().rename(columns={'Sales':'AvgMonthlySales'})
avg_monthly['MonthlyTarget'] = avg_monthly['AvgMonthlySales'] * 1.10   # 10% uplift target
# merge back (if you want a per-row target)
data = data.merge(avg_monthly[['Region','MonthlyTarget']], on='Region', how='left')
```

```python
assumed_cr = 0.20
data['EstimatedPipelineNeeded'] = data['Sales'] / assumed_cr
```
